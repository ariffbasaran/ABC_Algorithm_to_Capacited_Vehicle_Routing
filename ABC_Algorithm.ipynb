```python
#BESİN MATRİSİ OLUŞTURMA
def besinmatrisiolustur(besinsayisi, nodesayisi):
    # Boş listeleri içeren bir liste oluşturalım
    listeler = []

    for _ in range(besinsayisi):  # 5 tane liste oluşturmak için döngü
        liste = []

        while len(liste) != nodesayisi:
            try:
                if len(liste) == 0:
                    liste.append(0)
                elif len(liste) > 0 and len(liste) < sonnode:
                    i = liste[-1]
                    if (i % 2 == 0) and (i < ((nodesayisi-2) / 2)):
                        rastgele_cift = random.choice([num for num in range(2, nodesayisi, 2) if num not in liste])
                        liste.append(rastgele_cift)
                    elif (i % 2 != 0) and (i >= ((nodesayisi-2) / 2)):
                        rastgele_tek = random.choice([num for num in range(1, nodesayisi, 2) if num not in liste and num != sonnode])
                        liste.append(rastgele_tek)
                    elif (i % 2 == 0) and (i >= ((nodesayisi-2) / 2)):
                        rastgele_eleman = random.choice([num for num in range(0, nodesayisi) if num not in liste and num != sonnode])
                        liste.append(rastgele_eleman)
                    elif (i % 2 != 0) and (i < ((nodesayisi-2) / 2)):
                        rastgele_eleman = random.choice([num for num in range(0, nodesayisi) if num not in liste and num != sonnode])
                        liste.append(rastgele_eleman)
                elif len(liste) == sonnode:
                    liste.append(sonnode)
            except:
                liste = []  # Hata oluştuğunda liste sıfırlanacak, başa dönecek.

        listeler.append(liste)  # Oluşturulan liste listelere eklenecek

    besinler = np.array(listeler)
    return besinler
```

```python
#YER DEGİSTİRME
def yerdegistir(besin, indis1, indis2):
    yedekbesin  = besinler[besin].copy()
    islem_kontrol = 0
    if (indis1 == 0) or (indis2 == sonnode):
      #print("islem yapilamaz")
      islem_kontrol=0
      return yedekbesin, islem_kontrol
    else:

      yenibesin =  besinler[besin].copy()
      bosdegisken = yenibesin[indis1]
      yenibesin[indis1] = yenibesin[indis2]
      yenibesin[indis2] = bosdegisken

      for i in range(0,len(besinler[besin])-1):
        if (yenibesin[i] %2 == 0) and ((yenibesin[i+1])%2 !=0) and (yenibesin[i]< ((nodesayisi-2)/2)):
            #print("islem yapilamaz")
            islem_kontrol=0
            return yedekbesin, islem_kontrol
        elif (yenibesin[i] %2 != 0) and ((yenibesin[i+1])%2 ==0) and (yenibesin[i]>= ((nodesayisi-2)/2)):
            #print("islem yapilamaz")
            islem_kontrol=0
            return yedekbesin, islem_kontrol
        else:
            islem_kontrol  = 1
            if (yenibesin[0] != 0) or (yenibesin[-1] != sonnode):
              islem_kontrol=0
              return yedekbesin, islem_kontrol
            else:
              tekrar_ediyor = any(count > 1 for count in Counter(yenibesin).values())
              if tekrar_ediyor:
                islem_kontrol=0
                return yedekbesin, islem_kontrol

    return yenibesin, islem_kontrol
```

```python
#SOLA KAYDRIMA

def solakaydir(besin, indis1, indis2):
    yedekbesin  = besinler[besin].copy()
    islem_kontrol = 0
    if (indis1 == 0) or (indis2 == sonnode):
      #print("islem yapilamaz")
      islem_kontrol=0
      return yedekbesin, islem_kontrol
    else:
      yenibesin =  besinler[besin].copy()
      kucuk = min(indis1, indis2)
      buyuk = max(indis1, indis2)
      ilkeleman = yenibesin[kucuk]

      for i in range(kucuk, buyuk-1):
        yenibesin[i] = yenibesin[i+1]
      yenibesin[buyuk] = ilkeleman;

      for i in range(0,len(besinler[besin])-1):
        if (yenibesin[i] %2 == 0) and ((yenibesin[i+1])%2 !=0) and (yenibesin[i]< ((nodesayisi-2)/2)):
            #print("islem yapilamaz")
            islem_kontrol=0
            return yedekbesin, islem_kontrol
        elif (yenibesin[i] %2 != 0) and ((yenibesin[i+1])%2 ==0) and (yenibesin[i]>= ((nodesayisi-2)/2)):
            #print("islem yapilamaz")
            islem_kontrol=0
            return yedekbesin, islem_kontrol
        else:
            islem_kontrol  = 1
            if (yenibesin[0] != 0) or (yenibesin[-1] != sonnode):
              islem_kontrol=0
              return yedekbesin, islem_kontrol
            else:
              tekrar_ediyor = any(count > 1 for count in Counter(yenibesin).values())
              if tekrar_ediyor:
                islem_kontrol=0
                return yedekbesin, islem_kontrol

    return yenibesin, islem_kontrol

```

```python
#DİZİYİ TERS ÇEVİR OPERATORU
def diziyiterscevir(besin,indis1,indis2):
   yedekbesin  = besinler[besin].copy()
   islem_kontrol = 0
   if (indis1 == 0) or (indis2 == sonnode):
      #print("islem yapilamaz")
      islem_kontrol=0
      return yedekbesin, islem_kontrol
   else:
      yenibesin =  besinler[besin].copy()
      kucuk = min(indis1, indis2)
      buyuk = max(indis1, indis2)
      ilkeleman = yenibesin[kucuk]
      while kucuk<buyuk:
        bosdegisken = yenibesin[kucuk]
        yenibesin[kucuk] = yenibesin[buyuk]
        yenibesin[buyuk] = bosdegisken
        kucuk = kucuk+1
        buyuk = buyuk -1
      for i in range(0,len(besinler[besin])-1):
        if (yenibesin[i] %2 == 0) and ((yenibesin[i+1])%2 !=0) and (yenibesin[i]< ((nodesayisi-2)/2)):
            #print("islem yapilamaz")
            islem_kontrol=0
            return yedekbesin, islem_kontrol
        elif (yenibesin[i] %2 != 0) and ((yenibesin[i+1])%2 ==0) and (yenibesin[i]>= ((nodesayisi-2)/2)):
            #print("islem yapilamaz")
            islem_kontrol=0
            return yedekbesin, islem_kontrol
        else:
            islem_kontrol  = 1
            if (yenibesin[0] != 0) or (yenibesin[-1] != sonnode):
              islem_kontrol=0
              return yedekbesin, islem_kontrol
            else:
              tekrar_ediyor = any(count > 1 for count in Counter(yenibesin).values())
              if tekrar_ediyor:
                islem_kontrol=0
                return yedekbesin, islem_kontrol

   return yenibesin, islem_kontrol
```
